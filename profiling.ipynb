```python
data = fetch_openml(name="Boston", as_frame=True)
data
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df["TARGET"] = data.target
df
```

```python
profile = ProfileReport(df)
profile.to_file(output_file="boston_housing_profile.html")
```
